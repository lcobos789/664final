```python
import pandas as pd

csv_path = 'nationalgallery_objects.csv'

df = pd.read_csv(csv_path)

indigenous_count = df.apply(lambda row: 'indigenous' in str(row).lower(), axis=1).sum()

print(f'The number of indigenous objects is: {indigenous_count}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table
import random

csv_path = 'nationalgallery_objects.csv'


df = pd.read_csv(csv_path)

indian_in_title_objects = df[df['title'].str.lower().str.contains('indian')]

random_responses = indian_in_title_objects.sample(n=10)

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
tab = table(ax, random_responses[['title', 'attribution']], loc='center', colWidths=[0.3]*len(random_responses.columns))
tab.auto_set_font_size(False)
tab.set_fontsize(10)
tab.scale(3, 1.2) 

plt.title('Objects with "Indian" in Title - 10 Random Samples')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = 'nationalgallery_objects.csv'


df = pd.read_csv(csv_path)

keywords = ['japan', 'china', 'india', 'mexico','peru']

keyword_counts = {word: df['title'].str.lower().str.count(word).sum() for word in keywords}

plt.bar(keyword_counts.keys(), keyword_counts.values())
plt.xlabel('Keywords')
plt.ylabel('Count')
plt.title('Count of Keywords in Title')
plt.show()
```
